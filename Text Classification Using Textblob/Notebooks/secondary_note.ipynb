```python
import pandas as pd
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df_apple = pd.read_csv(r"D:\Files\DSC\Capstone Project 1\Datasets\apple_dataset.csv",index_col=0)
df_apple.head()
```

```python
df_samsung = pd.read_csv(r"D:\Files\DSC\Capstone Project 1\Datasets\samsung_dataset.csv",index_col=0)
df_samsung.head()
```

```python
#ploting top 5 coutries from where the tweets originated.

plot, ax = plt.subplots(1,2,constrained_layout=True,figsize=(10,5))
sns.set(style='darkgrid')
sns.countplot(y='country',data=df_apple,order=df_apple.country[df_apple.country!='Unknown'].value_counts().iloc[:5].index,ax=ax[0])
sns.countplot(y='country',data=df_samsung,order=df_samsung.country[df_samsung.country!='Unknown'].value_counts().iloc[:5].index,ax=ax[1])
ax[0].title.set_text('Apple')
ax[1].title.set_text('Samsung')
plot.show()
```

```python
# The suprising thing here is we see India in Top 5 tweets for Apple. And that is likely to hold true becacuse of large
# number of showroom deals and bank offers. As of Nov 2019 the upper segment of phones (Rs 35000 and above) Apple 
# holds 50 % of the market share pushing OnePlus to the 2nd spot. And overall Apple is now in top 10 smartphone brands.
```

```python
# nltk.download()  # to be used if copra has not been downloaded
sws_apple = stopwords.words('english')
sws_samsung = stopwords.words('english')
```

```python
sws_apple.extend(['iphone','iPhone','IPHONE','Apple','Phone','i Phone','iphone11','iPhone11','iPhone 11','iPhone11Pro','new','iPhone11ProMax'])
sws_samsung.extend(['Samsung','samsung','Galaxy','GalaxyS10','Galaxys10','galaxys10'])
```

```python
txt = ' '.join(df_apple.tweet)
stopword = set(STOPWORDS)
stopword = (sws_apple)
wc = WordCloud(background_color='white',stopwords=stopword).generate(txt)
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()

# Most common words/things/features that people associate after launch of iphone 11
```

```python
txt = ' '.join(df_samsung.tweet)
stopword = set(STOPWORDS)
stopword = (sws_samsung)
wc = WordCloud(background_color='white',stopwords=stopword).generate(txt)
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# We can see how people have described (liked features) for Apple.
#Smart Battery (ability of the phone to learn ur usage and optimize battery),Battery Case, Pay options, Max and Pro versions

#for Samsung its
# Displays(home and lock screen),Gradient phone body,android OS,One UI
```

```python
#df_

plot, ax = plt.subplots(1,2,constrained_layout=True,figsize=(10,5))
sns.set(style='darkgrid')
sns.barplot(x="sentiment", y="retweet_count", data=df_apple, estimator=lambda x: len(x) / len(df_apple) * 100,ax=ax[0])
sns.barplot(x="sentiment", y="retweet_count", data=df_samsung, estimator=lambda x: len(x) / len(df_samsung) * 100,ax=ax[1])
ax[0].title.set_text('Apple')
ax[1].title.set_text('Samsung')
plot.show()
```
