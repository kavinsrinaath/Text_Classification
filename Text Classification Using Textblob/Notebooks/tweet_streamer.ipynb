```python
# parsing the tweets into json format

tweets_file = 'fetched_tweets.txt'

tweet_data = open(tweets_file,'r')
tweet_arry = []
for line in tweet_data:
    try:
        tweet = json.loads(line)
        tweet_arry.append(tweet)
    except:
        continue
```

```python
# loading the data into apndas and piciking out only text and remove handle names, special chars, emojis
df = pd.DataFrame(tweet_arry)
df = df[['text']]
df['text'] = data_cleaner_tweets(df.text)
```

```python
# Filtering out tweets of iphone and galaxy. Using indices to match any tweet that has both mentioned

apple_index = set(df[df.text.str.contains('iphone|apple',case=False)].index)
samsung_index=(df[df.text.str.contains('galaxy|samsung',case=False)].index)

common = apple_index.intersection(samsung_index)
apple_index = apple_index.difference(common)
samsung_index = samsung_index.difference(common)
```

```python
# Getting the number of tweets for each category
apple_tweets = len(apple_index)
samsung_tweets = len(samsung_index)
common_tweets = len(common)
```

```python
#pasrsing the data into a dataframe for easier visualization using seaborn

raw_data = {'tweet_mentions':['Apple','Samsung','Common'],'no_of_tweets':[apple_tweets,samsung_tweets,common_tweets]}
df_main = pd.DataFrame(raw_data)
```

```python
sns.set(style='darkgrid')
sns.barplot(x='tweet_mentions',y='no_of_tweets',data=df_main)
```
